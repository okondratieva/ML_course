#### Основные методы классов, реализующих алгоритмы машинного обучения
Все алгоритмы выполнены в виде классов, обладающих по крайней мере следующими методами:



   Метод класса            |  Описание     
 --------------------------|-------------------------------------------------------------
  `fit(X_train, y_train)`              |  обучение (тренировка) модели на обучающей выборке `X_train`, `y_train`
  `predict(X)`             |  предсказание на данных `X`
  `set_params(**params)`   |  установка параметров алгоритма
  `get_params()`           |  чтение параметров алгоритма
 
 Заметим, что параметры алгоритмов обучения можно задавать как в конструкторе класса, так и с помощью метода `set_params(**params)`.
 
 #### Таблица наличия/отсутствия методов для основных алгоритмов машинного обучения:

### Подбор оптимального alpha 

```python
alpha_arr = np.logspace(0, 3, 21)#возвращает числа распределенные равномерно на логарифмической шкале, в данном случае 21 число 
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (50,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-*', label = 'train')
plt.semilogx(alpha_arr, test_err, 'y-o', label = 'test')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

Найдем минимальные значения ошибки в обучающей и тестовой выборках

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
alpha_opt = alpha_arr[test_err == min_test_err] #оптимальное значение alpha
print(alpha_opt)
```

```python
alpha_opt = alpha_opt[0]
```

```python
alpha_opt
```

### Классификатор с оптимальным alpha

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (50,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

Посмотрим зависимость величины ошибки от числа нейронов

```python
hidden_layers_sizes = range(1, 50, 1)
test_err = []
train_err = []

for layer_size in hidden_layers_sizes:
    mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (layer_size, ), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.plot(hidden_layers_sizes, train_err, 'b-o', label = 'train')
plt.plot(hidden_layers_sizes, test_err, 'y-o', label = 'test')
plt.xlim([np.min(hidden_layers_sizes), np.max(hidden_layers_sizes)])
plt.title('Error vs. hidden layers size')
plt.xlabel('hidden layers size')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
#ищем минимальное количество нейронов, необходимое для сети для получения минимальной ошибки
for i in range(len(test_err)):
    if(test_err[i] == min_test_err):
        hidden_layers_size_opt = hidden_layers_sizes[i]
        break
print(hidden_layers_size_opt)
```

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (hidden_layers_size_opt,),
                          solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
print(np.mean(y_train != y_train_pred))
print(np.mean(y_test != y_test_pred))
```

## Random Forest

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

Далее проведем эксперименты с различным числом деревьев (n_estimators) или классификаторов для каждого из алгоритмов обучения

```python
trees_count=range(1,100,1)
test_err = []
train_err = []
oob_err=[]
for tree in trees_count:
    #изменим максимальную глубину деревьев, чтобы избежать переобучения
    rf = ensemble.RandomForestClassifier(n_estimators = tree, max_depth=7,random_state=7, oob_score=True)
    rf.fit(X_train, y_train)
    y_train_pred = rf.predict(X_train)
    y_test_pred = rf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - rf.oob_score_)
print(tree)
```

```python
plt.plot(trees_count, train_err, 'b-o', label = 'train')
plt.plot(trees_count, test_err, 'g-o', label = 'test')
plt.plot(trees_count, oob_err, 'r-o', label = 'oob')
plt.xlim([np.min(trees_count), np.max(trees_count)])
plt.title('Error vs. count of trees')
plt.xlabel('counts of trees')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
#ищем минимальное количество деревьев, необходимое для получения минимальной ошибки
for i in range(len(test_err)):
    if(test_err[i] == min_test_err):
        trees_opt = trees_count[i]
        break
print(trees_opt)
```

```python
rf = ensemble.RandomForestClassifier(n_estimators = trees_opt, max_depth=7, random_state=7)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

## Extremely Random Forests

```python
erf=ensemble.ExtraTreesClassifier(n_estimators=50).fit(X_train, y_train)
err_train = np.mean(y_train != erf.predict(X_train))
err_test = np.mean(y_test != erf.predict(X_test))

print(err_train, err_test)
```

```python
trees_count=range(1,100,1)
test_err = []
train_err = []
oob_err=[]

for tree in trees_count:
    #изменим максимальную глубину деревьев, чтобы избежать переобучения  
    erf = ensemble.ExtraTreesClassifier(n_estimators = tree, max_depth=7, random_state=7, oob_score=True, bootstrap=True).fit(X_train, y_train)
    y_train_pred = erf.predict(X_train)
    y_test_pred = erf.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
    oob_err.append(1 - erf.oob_score_)
plt.plot(trees_count, train_err, 'b-o', label = 'train')
plt.plot(trees_count, test_err, 'g-o', label = 'test')
plt.plot(trees_count, oob_err, 'r-o', label = 'oob')
plt.xlim([np.min(trees_count), np.max(trees_count)])
plt.title('Error vs. count of trees')
plt.xlabel('counts of trees')
plt.ylabel('error')
plt.legend()
pass
```

Неограниченная глубина деревьев может привести к переобучению, поэтому изменим параметр max_depth максимальной глубины

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
#ищем минимальное количество деревьев, необходимое для получения минимальной ошибки
for i in range(len(test_err)):
    if(test_err[i] == min_test_err):
        trees_opt = trees_count[i]
        break
print(trees_opt)
```

```python
erf = ensemble.ExtraTreesClassifier(n_estimators = trees_opt, max_depth=4, random_state=7).fit(X_train, y_train)
y_train_pred = erf.predict(X_train)
y_test_pred = erf.predict(X_test)
err_train=np.mean(y_train != y_train_pred)
err_test=np.mean(y_test != y_test_pred)
print(err_train, err_test)
```

```python
components = range(1, 28)
train_err = []
test_err = []
for c in components:
    pca = PCA(n_components = c)
    pca.fit(X)
    x_pca = pca.transform(X)
    X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(x_pca, y, test_size = 0.33, random_state = 42)
    mlp.fit(X_train_pca, y_train_pca)
    train_err.append(np.mean(y_train_pca != mlp.predict(X_train_pca)))
    test_err.append(np.mean(y_test_pca != mlp.predict(X_test_pca)))
    
plt.plot(components, train_err, 'b-o', label = 'train')
plt.plot(components, test_err, 'r-o', label = 'test')
plt.xlim([np.min(components), np.max(components)])
plt.title('Error vs. n_components')
plt.xlabel('n_components')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)
print(min_train_err, min_test_err)
```

```python
for i in range(len(test_err)):
    if(test_err[i] == min_test_err):
        components_opt = components[i]
        break
print(components_opt)
```
